```python
def checkValidPlacement(sudoku, row, col, val):
    #check row
    if(val in sudoku[row]):
        #print("Failed row: " + str(row))
        return False

    #check col
    for i in range(len(sudoku)):
        if(val == sudoku[i][col]):
            #print("Failed col: " + str(i) + ", " + str(col))
            return False

    #Check square
    #Get section of square row 0-2, 3-5, 6-9 (1-3, 4-6, 7-9)
    rowSectionStart = 0
    rowSectionEnd = 0
    if(row <= 2):
        rowSectionStart = 0
        rowSectionEnd = 2
    elif((row >= 3) and (row <= 5)):
        rowSectionStart = 3
        rowSectionEnd = 5
    else: #row > 6
        rowSectionStart = 6
        rowSectionEnd = 8

    #Get section of square col 0-2, 3-5, 6-9 (1-3, 4-6, 7-9)
    colSectionStart = 0
    colSectionEnd = 0
    if(col <= 2):
        colSectionStart = 0
        colSectionEnd = 2
    elif((col >= 3) and (col <= 5)):
        colSectionStart = 3
        colSectionEnd = 5
    else: #row > 6
        colSectionStart = 6
        colSectionEnd = 8

    #check the square
    for i in range(rowSectionStart, rowSectionEnd + 1):
        for k in range(colSectionStart, colSectionEnd + 1):
            if(val == sudoku[i][k]):
                #print("Failed square: " + str(i) + ", " + str(k))
                return False
    
    return True
```

```python
#TODO:Check if puzzle is complete

#TODO:Backtracking
def sudokuSolveBacktracking(sudoku, row, col, endCondition):

    #end Case
    if((row == 9) and (col == 9)):
        #puzzle complete Check if it is valid
        #print("=======================================Completed!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!")
        endCondition = True
        #print("Returning.......................................................")
        return sudoku, endCondition
        #print("after return")
    
    elif(col == 9):
        #print("Row completed: " + str(row))
        if(row < 8):
            #Move to next row
             sudoku, endCondition =sudokuSolveBacktracking(sudoku, row + 1, 0, endCondition)
        else:
            #Finished last row
            sudoku, endCondition =sudokuSolveBacktracking(sudoku, row + 1, 9, endCondition)
    
    elif(sudoku[row][col] != '.'):
        #We can skip this position since it is already filled
        sudoku, endCondition = sudokuSolveBacktracking(sudoku, row, col + 1, endCondition)

    else:
        #Loop through all possible values and check end condition 
        #print("Row completed: " + str(row))
        i = 1
        while((i <= 9) and (endCondition == False)):
            if(checkValidPlacement(sudoku, row, col, i)):
                sudoku[row][col] = i
                #print("Row: "+ str(row) + ", col: " + str(col) + ", Valid: " + str(i))
                #print(sudoku)
                sudoku, endCondition = sudokuSolveBacktracking(sudoku, row, col + 1, endCondition)
                #if(endCondition == True):
                    #print("Endcondition is true end backtracking")
                #print("i: " + str(i) + ", endCondition: " + str(endCondition))
            i = i + 1
        
        #if when through all the options and cant place then when backtracking takeout the value used
        if((i >= 9) and (endCondition == False)):
            sudoku[row][col] = '.'
            #print("i >= 9, Erase and backtrack")
    
    #print("Return")
    return sudoku, endCondition
        
```

```python
def printSudokuPuzzle(sudokuPuzzle):
    for row in range(len(sudokuPuzzle)):
        for col in range(len(sudokuPuzzle[row])):
            print(str(sudokuPuzzle[row][col]) + "\t", end='')
        print()
```
